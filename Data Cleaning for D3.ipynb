```python
DFXXX = df[["WEATHER_DELAY",'DEPHum','DEPPres','DEPTemp','DEPWD','DEPWS','ARRHum','ARRPres','ARRTemp','ARRWD','ARRWS','mon']]
```

```python
DFXXX = DFXXX[DFXXX["WEATHER_DELAY"]>=0]
DFXXX["mon"] = DFXXX["mon"].apply(int)
DFXXX.corr()
```

```python
peak = {12, 1, 2, 6, 7, 8}
```

```python
DFXXX["Peak"] = (DFXXX["mon"] ==12) |(DFXXX["mon"] ==1) |(DFXXX["mon"] ==2)|(DFXXX["mon"] ==6)| (DFXXX["mon"] ==7) |(DFXXX["mon"] ==8)
DFXXX["Peak"] = DFXXX["Peak"].apply(int)
```

```python
DFXXX["Delayed"] = DFXXX["WEATHER_DELAY"] >0
DFXXX["Delayed"] = DFXXX["Delayed"].apply(int)
DFXXX["Delayed"]
```

```python
from sklearn.model_selection import train_test_split
X = DFXXX[["DEPHum","DEPPres","DEPTemp", "DEPWS", "ARRHum", "ARRTemp", "ARRWS", "Peak"]]
y = DFXXX['Delayed']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtree = DecisionTreeRegressor()
dtree.fit(X_train,y_train)
predictions = dtree.predict(X_test)
```

```python
dtree.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
rtree = RandomForestClassifier()
rtree.fit(X_train,y_train)
predictions = rtree.predict(X_test)
```

```python
rtree.score(X_test, y_test)
```
